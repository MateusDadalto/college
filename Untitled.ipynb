```python
import mixed_current
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
result = []

x_axis = [i for i in range(10,41,1)]

for pi_c in x_axis:
    result.append(mixed_current.calculate(pi_c, 2, 0.9, False)[1] * 1000 * 3600)

```

```python
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(111)

M_0s = [0.9,2]
pi_fs = [2,3]

for pi_f in pi_fs:
    for M in M_0s:
        result = []
        for pi_c in x_axis:
            result.append(mixed_current.calculate(pi_c, pi_f, M, False)[1] * 1000 * 3600)
        
        ax.plot(x_axis, result, label='pi_f: {0} | M: {1}'.format(pi_f, M))

plt.legend(loc=1)
plt.ylabel('S [kg/(h*kN)]')
```
