```python
import matplotlib.pyplot as plt
import pandas as pd
from nozzle import cf, P_c, A_t
```

```python
df = pd.read_csv('usa_atmosphere')
df
```

```python
traction = []
for p0 in df['press']: 
    p0 = p0/1E5
    t = cf(p0) * P_c * 1E5 * A_t
    traction.append(t/9.81/1000)
    
traction
```

```python
fig = plt.figure(figsize=(10,10))
plt.plot(df['alt'], traction)
```
